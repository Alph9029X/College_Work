```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
n=1000

data={
    'Date': pd.date_range(start='2020-01-01',periods=n),
    'Category': [random.choice(categories)for _ in range(n)],
    'Likes': np.random.randint(0,20000,size=n)
}
```

```python
df=pd.DataFrame(data)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

# to count each unique element in 'Category'

```python
category_counts=df['Category'].value_counts()
print(category_counts)
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Likes']=df['Likes'].astype(int)
```

```python
print(df)
```

```python
sns.histplot(data=df, x='Likes')
plt.title("Histogram of likes")
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df, x='Category', y='Likes')
plt.title('Boxplot of Likes by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
mean_likes=df['Likes'].mean()
print(f"Mean Likes: {mean_likes: .2f}")
```

```python
category_mean_likes=df.groupby('Category')['Likes'].mean()
print("\nMean Likes by Category:")
print(category_mean_likes)
```
